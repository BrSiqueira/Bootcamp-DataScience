##Desafio 03
###Formatar o gráfico de custos por mês dos 5 estados, deixando ele agradável(Bonitão, segundo o Gui)

Seleciona os estados que mais gastaram

```python
ordenado_total = usaveis
ordenado_total["Total"] = ordenado_total.sum(axis=1).copy()
ordenado_total = ordenado_total.sort_values("Total", ascending=False)
ordenado_total = ordenado_total.drop("Total", axis=1)
ordenado_total.head(5)
```

```python
ax = ordenado_total.iloc[:5, 8:].T.plot(figsize=(19,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(loc="upper center", mode = "expand", ncol = 7, bbox_to_anchor=(0, -1.05, 1 , 1))
plt.title('Filtrando 5 estados -v1', fontsize=20)
plt.show()
```
