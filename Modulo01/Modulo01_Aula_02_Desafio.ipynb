## Aula 01, Pate 2

### Desafio 1: Melhorar o título do gráfico

```python
import matplotlib.ticker as ticker
ax = dados.plot(x="Unidade da Federação", y=["2008/Ago", "2009/Ago"], kind="bar", figsize=(18,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title('Comparação dos gastos de 2008 e 2009')
plt.show()
```

```python
ax = dados.plot(x="Unidade da Federação", y=[ "2019/Mar", "2020/Mar", "2021/Mar"], kind="bar", figsize=(18,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title('Comparação dos Meses mais recentes')
plt.show()
```

```python
ax = dados.plot(x="Unidade da Federação", y=dados.filter(regex="Mar").columns, figsize=(18,8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title('Comparação dos Meses mais recentes')
plt.show()
```

```python
s1 = dados.filter(regex = 'Mar')
s2 = dados['Unidade da Federação']
s3 = pd.concat([s2, s1], axis=1)

s3.set_index('Unidade da Federação', drop=True, inplace = True)
s3.T
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
# usaveis = usaveis.set_index('Unidade da Federação')
usaveis.set_index('Unidade da Federação', inplace=True)
usaveis
```

```python
usaveis['2019/Ago']
```

```python
usaveis.loc["12 Acre"]
```

```python
usaveis = usaveis.drop("Total", axis=1)
```

```python
usaveis.T.plot(figsize=(19,6))
```

##Desafio 1
###Retirar a legenda de dentro do gráfico


```python
ax = usaveis.T.plot(figsize=(19,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(loc="upper center", mode = "expand", ncol = 7, bbox_to_anchor=(0, -1.05, 1 , 1))
plt.title('Comparação dos Meses mais recentes')
plt.show()
```

```python
ax = dados.sort_values("Total",ascending=False).plot.bar(x="Unidade da Federação", y="Total", figsize=(19,6))
plt.show()
```
